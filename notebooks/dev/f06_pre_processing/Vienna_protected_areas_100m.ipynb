```python
## pa raster allready re-projected:


pa_10m_raster = r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\10m_GRID\vienna_schutzgebiete_10m.tif"

## reference poly: 100x100m:
reference_polygon = r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\reference_units\GRID\grid31256\fishnet_100m_31256.shp"

## output:
out_pa2018_100m_vector =r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\100m_GRID\Wien_protected_areas_by_grid100m.shp"
out_pa2018_100m_vector_indicator=r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\100m_GRID\Wien_protected_areas_by_grid100m_indicator.shp"


out_pa2018_100m_vector_area=r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\100m_GRID\Wien_protected_areas_by_grid100m_area.shp"
```

```python

## zonal statistics: output = vector dataset
import geopandas as gpd
from rasterstats import zonal_stats

zones = reference_polygon
values = pa_10m_raster
gdf = gpd.read_file(zones)
print(gdf.head())

stats = gpd.GeoDataFrame(zonal_stats(gdf, values, stats=["min", "max", "mean",'sum']))
gdf = gdf.join(stats)

print(gdf.head())


gdf.to_file(out_pa2018_100m_vector)
print ("done")
```

produce a 100m tif from gridded vector dataset:

## normalization on vector data & calc. percent per cell:

```python
## area o pa inside 100m cells: -------------------------------------
column = 'sum'
gdf['sum'] = gdf['sum'].fillna(0)   # set null to "0"
gdf['out_pa2018_100m_area'] = (gdf[column] )	 
gdf.to_file(out_pa2018_100m_vector_area)


## indicator: -------------------------------------
# apply normalization techniques by Column 1  (0-100)
column = 'sum'
gdf['out_pa2018_100m_vector'] = (gdf[column] - gdf[column].min()) / (gdf[column].max() - gdf[column].min())	 *100
# view normalized data 
display(gdf) 

gdf.to_file(out_pa2018_100m_vector_indicator)

print ("done")
```

# Export vector data to raster:

```python

## 1 export area values raster:

## raster to vector using vector attriubte (working perfectly):

import geopandas as gpd
import rasterio
from rasterio.features import rasterize
from rasterio.transform import from_bounds

# Load the polygon layer
polygons = gpd.read_file(out_pa2018_100m_vector_indicator)

output_raster = r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\100m_GRID\r06_protected_area_100m.tif"  


# Define the attribute to use for raster values
attribute_to_rasterize = 'sum'

# Define raster properties
#xmin, ymin, xmax, ymax = polygons.total_bounds

xmin=	-11200
ymin=	331000
xmax=	18800
ymax=	354000

projection ='epsg:31256'


pixel_size = 100  # Adjust as needed


width = int((xmax - xmin) / pixel_size)
height = int((ymax - ymin) / pixel_size)
transform = from_bounds(xmin, ymin, xmax, ymax, width, height)

# Create an empty array to hold rasterized values
raster_array = rasterize(
    [(geom, value) for geom, value in zip(polygons.geometry, polygons[attribute_to_rasterize])],
    out_shape=(height, width),
    transform=transform,
    fill=0,
    all_touched=False,  # or True
    dtype=rasterio.float32
)

# Write the raster to a GeoTIFF file
with rasterio.open(output_raster, "w", driver="GTiff", 
                   height=raster_array.shape[0], 
                   width=raster_array.shape[1], 
                   count=1, 
                   dtype=raster_array.dtype, 
                   crs=polygons.crs, 
                   transform=transform
                   ) as dst:
    dst.write(raster_array, 1)


print ("area raster with ha values done")

###########################

##export indicator:

##

## raster to vector using vector attriubte (working perfectly):

import geopandas as gpd
import rasterio
from rasterio.features import rasterize
from rasterio.transform import from_bounds

# Load the polygon layer
polygons = gpd.read_file(out_pa2018_100m_vector_indicator)

output_raster = r"N:\C2205_FAIRiCUBE\f02_data\d050_data_WIEN\Vienna_gridded_data_collection\100m_GRID\r06_protected_areas_indicator0100.tif"  


# Define the attribute to use for raster values
attribute_to_rasterize = 'out_pa2018'

# Define raster properties
#xmin, ymin, xmax, ymax = polygons.total_bounds

xmin=	-11200
ymin=	331000
xmax=	18800
ymax=	354000

projection ='epsg:31256'


pixel_size = 100  # Adjust as needed


width = int((xmax - xmin) / pixel_size)
height = int((ymax - ymin) / pixel_size)
transform = from_bounds(xmin, ymin, xmax, ymax, width, height)

# Create an empty array to hold rasterized values
raster_array = rasterize(
    [(geom, value) for geom, value in zip(polygons.geometry, polygons[attribute_to_rasterize])],
    out_shape=(height, width),
    transform=transform,
    fill=0,
    all_touched=False,  # or True
    dtype=rasterio.float32
)

# Write the raster to a GeoTIFF file
with rasterio.open(output_raster, "w", driver="GTiff", 
                   height=raster_array.shape[0], 
                   width=raster_array.shape[1], 
                   count=1, 
                   dtype=raster_array.dtype, 
                   crs=polygons.crs, 
                   transform=transform
                   ) as dst:
    dst.write(raster_array, 1)


print ("indicator raster done")
```
