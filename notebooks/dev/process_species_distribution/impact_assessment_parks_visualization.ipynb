```python
# draw horizontal barplot with label park name and value mean
# sort values by mean
df = df[:10]
df.sort_values(by='_mean', inplace=True, ascending=True)
df.plot.barh(x='park_name', y='_mean')
# add lines indicating standard deviation
plt.errorbar(df['_mean'], df.park_name, xerr=df['_stdev'], fmt='.',  color='black', capsize=2)
# hide spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
# change plt height
plt.gcf().set_size_inches(6, 8)
plt.tick_params(axis=u'both', which=u'both',length=0)
# remove legend
plt.legend().set_visible(False)
# remove ylabel
plt.ylabel('')
# increase font size of x-axis label
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
merged_df = pd.read_csv("N:/C2205_FAIRiCUBE/f02_data/d060_data_LUXEMBOURG/f05_distribution/f03_distribution_maxent/zonal_statistics/parks_zonal_stats_all.csv")
merged_df = merged_df[(merged_df._count > 100)]
merged_df.sort_values(by='_count', ascending=False, inplace=True)
```

```python
merged_df
```

```python
import seaborn as sns
# Extract just the mean values for the heatmap
mean_columns = [col for col in merged_df.columns if col.endswith('_mean')]
# compute average of all mean values
heatmap_data = merged_df[['park_name'] + mean_columns].set_index('park_name')
```

```python
for idx in heatmap_data.index:
    print(idx)
```

```python

heatmap_data['mean'] = heatmap_data.mean(axis=1)
heatmap_data.sort_values(by="mean", ascending=False, inplace=True)
heatmap_data = heatmap_data.drop(columns=['mean'])
heatmap_data.columns = ['Robinia pseudoacacia', 
                        'Heracleum mantegazzianum',
                        'Solidago canadensis', 
                        'Prunus laurocerasus', 
                        'Impatiens glandulifera',
                        'Reynoutria japonica']

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title('Mean suitability index across species and Parks', fontsize=16)
plt.ylabel('Municipal Parks', fontsize=12)
plt.xlabel('Species', fontsize=12)
# show xlabel on top
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
# show xlabel with 45 degree rotation
plt.xticks(rotation=15, ha='left')
# save the plot
plt.savefig("N:/C2205_FAIRiCUBE/f07_images/VdL_neophytes/parks_zonal_stats_all.png", dpi=300)
# plt.tight_layout()
plt.show()
```
