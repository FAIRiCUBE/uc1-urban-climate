```python
import rasterio
from matplotlib import pyplot
dataset = rasterio.open(r'UrbanAtlas2018_lux.tif')
pyplot.imshow(dataset.read(1), cmap='pink')
```

```python
dataset.statistics(True)
```

Count the number of points for each typology

```python
import numpy as np
typology = dataset.read(1)
a = np.array(typology)
unique, counts = np.unique(a, return_counts=True)
dict(zip(unique, counts))
nb_points = typology.size - counts[0]
```

```python
unique = np.delete(unique, 0)
counts = np.delete(counts, 0)
ratios = counts*100 / nb_points
```
