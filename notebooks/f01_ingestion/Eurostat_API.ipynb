````python
import requests
import json
import eurostat
````

````python
# list datasets in the theme Cities and greater cities
toc_df = eurostat.get_toc_df()
eurostat.subset_toc_df(toc_df, "cities and greater cities")
````

````python
pars = eurostat.get_pars('URB_CPOP1')
pars
````

````python
# list all variables
eurostat.get_dic("urb_cpopstr", "indic_ur")
````

````python
# get data for variable DE1001V, Population on the 1st of January, total
code = "URB_CPOP1"
filters = {"indic_ur": "DE1001V"}
df_urb_cpop1 = eurostat.get_data_df(code, flags=False, filter_pars=filters, verbose=True, reverse_time=False)
df_urb_cpop1
````
