# Data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in city UA l1 ratios
path = "./../../../s3/Machine Learning/city_UrbanAtlas_l1_ratios.csv"
df = pd.read_csv(path)
```

```python
city = df[df.cc_name == "Verona"]
city = pd.melt(city, id_vars=['cc_name'], value_vars=['Agricultural areas','Artificial surfaces','Natural and semi-natural areas',	'Water'])

myLabels = 'Agricultural areas','Artificial surfaces','Natural and semi-natural areas',	'Water',
plt.pie(city["value"],labels=myLabels,autopct='%1.2f%%')
plt.title('Verona')
plt.axis('equal')
plt.show()
```
